# ASSIGNMENT PROBABILITY BASICS
Question 1: A die is rolled. What is the probability of getting:

(a) An even number

(b) A number greater than 4

Sol :      

A die has outcome {1,2,3,4,5,6}

(a) Even numbers = {2,4,6} → 3 outcomes

Probability = 3/6 = 1/2

(b) Numbers greater than 4 = {5,6} → 2 outcomes

Probability = 2/6 = 1/3

Question 2: In a class of 50 students:

20 like Mathematics (M)

15 like Science (S)

5 like both subjects

What is the probability that a student chosen at random likes Mathematics or Science?

Sol :      
P(M ∪ S) = (M + S − Both) / Total

= (20 + 15 − 5) / 50

= 30/50

= 3/5

Question 3: A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability
that the next ball is also red (without replacement)?

Sol :       
After one red is drawn,

 remaining balls = 2 red, 2 blue (total 4)

Probability next ball is red = 2/4 = 1/2

Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?

Sol :      

 Stratified Sampling

Why?

Because the population has two groups (boys and girls), and stratified sampling ensures equal or proportional representation from each group.

Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height
= 158 cm.

Find the sampling error.

Sol :     
Sampling error = Sample mean − Population mean

= 158 − 160

 = −2 cm(Magnitude = 2 cm)

 Question 6: The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what
is the standard error of the mean (SEM)?

Sol :     
Formula for Standard Error:

SEM = 5000 / √100

= 5000 / 10

 = ₹500

Question 7: In a group of 100 students:

40 like Cricket (C)

30 like Football (F)

10 like both Cricket and Football

Find the probability that a student likes at least one sport.

Sol :      
P(C ∪ F) = (C + F − Both) / Total

= (40 + 30 − 10) / 100

= 60/100

= 3/5

Question 8: From a deck of 52 cards, two cards are drawn without replacement. What is the probability that
both are Aces?

Sol :     
Probability of 1st Ace = 4/52

Probability of 2nd Ace = 3/51

Total probability = (4/52) × (3/51)

= 12/2652

 = 1/221

Question 9: A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the
probability that all are non-defective?

Sol :      
Probability a bulb is non-defective = 1 − 0.02 = 0.98

Probability all 5 are non-defective =

P = (0.98)^5

≈ 0.9039 (approx)
